**3.8 Adding columns for "Total Score" and "Average"**

```python
df['total_score'] = df['math_score']+df['writing_score']+df['reading_score']
df['average_score'] = df['total_score']/3

df.head()
```

```python
math_full_marks = df[df['math_score'] == 100]['average_score'].count()
reading_score = df[df['reading_score'] == 100]['average_score'].count()
writing_score = df[df['writing_score'] == 100]['average_score'].count()
print(f'Number of student having full marks in Math {math_full_marks}')
print(f'Number of student having full marks in Reading {reading_score}')
print(f'Number of student having full marks in writing {writing_score}')
```

```python
math_full_marks = df[df['math_score'] <= 20]['average_score'].count()
reading_score = df[df['reading_score'] <= 20]['average_score'].count()
writing_score = df[df['writing_score'] <= 20]['average_score'].count()
print(f'Number of student having less than 20 marks in Math {math_full_marks}')
print(f'Number of student having less than 20 mark in Reading {reading_score}')
print(f'Number of student having less than 20 mark in writing {writing_score}')
```

**Insights**
* From above values we get students have performed the worst in Maths
* Best performance is in reading section

**4.4.4 LUNCH COLUMN**
* Which type of lunch is most common amoung students ?
* What is the effect of lunch type on test results?

**UNIVARIATE ANALYSIS ( Which type of lunch is most common amoung students ? )**

```python
plt.rcParams['figure.figsize'] = (15, 9)
plt.style.use('ggplot')
sns.countplot(df['lunch'], palette = 'PuBu')
plt.title('Comparison of different types of lunch', fontweight = 30, fontsize = 20)
plt.xlabel('types of lunch')
plt.ylabel('count')
plt.show()
```

**Insights**
* Students being served Standard lunch was more than free lunch

**BIVARIATE ANALYSIS ( Is lunch type intake has any impact on student's performance ? )**

```python
# Group by lunch and calculate mean of average score
lunch_avg = df.groupby('lunch')['average_score'].mean().reset_index()

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(data=lunch_avg, x='lunch', y='average_score', palette='Set2')

# Labels and title
plt.title('Average Score by Lunch Type', fontsize=16, fontweight='bold')
plt.xlabel('Lunch Type')
plt.ylabel('Average Score')
plt.ylim(0, 100)

# Show values on top of bars
for index, row in lunch_avg.iterrows():
    plt.text(index, row['average_score'] + 1, f"{row['average_score']:.1f}", ha='center', fontsize=12)

plt.show()
```

**Insights**
* Students who get Standard Lunch tend to perform better than students who got free/reduced lunch
